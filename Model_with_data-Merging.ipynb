```python
# Sample Data Run
from path import Path
import pandas as pd
import numpy as np
```

```python
data1 = Path('blue-team/Resources/Presidential_96_2020.csv')
df1 = pd.read_csv(data1)
df1.head()
```

```python
df1['Outcome'] = np.where(df1['Winning_Party']=='REPUBLICAN', 1, 0)
df1
```

```python
df1 = df1[df1.year >= 2000]
df1
```

```python
data2 = Path('blue-team/Resources/CENSUS_00_19.csv')
df2 = pd.read_csv(data2)
df2['STATE_NAME'] = df2['STATE_NAME'].str.upper()
df2
```

```python
merged_df = pd.merge(df1, df2, how='inner', left_on=['year','state'], right_on = ['YEAR','STATE_NAME'])
merged_df
```
