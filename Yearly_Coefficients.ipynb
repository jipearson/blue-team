```python
# Import Dependencies 
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import sqlalchemy as db
from sklearn import preprocessing
```

```python
from config import password
# Set up variables to connect to SQL
database = "presidential_migration_db"
user = "postgres_blue"
password = f'{password}'
host = "presidential-migration-db.cexjpewgnljl.us-east-2.rds.amazonaws.com"
port = "5432"
```

```python
# Create connection to Postgres DB using SQL Alchemy
engine = db.create_engine(f'postgresql://{user}:{password}@{host}:{port}/{database}')
engine.connect()
```

```python
# Pull in Election Demographics table from Postgres
df = pd.read_sql_query('SELECT * from election_dem', con=engine)
df.sort_values(by=['STATE', 'YEAR'], inplace=True)
df.set_index(['STATE', 'YEAR'], inplace=True)
df.drop(columns=['previous_senate_winning_party', 'electoral_votes'], inplace=True)
df
```

```python
# Scale data using MinMaxScaler
scaler = preprocessing.MinMaxScaler()
names = df.columns
index = df.index
d = scaler.fit_transform(df)

scaled_df = pd.DataFrame(d, columns=names, index=index)
scaled_df.reset_index(inplace=True)
scaled_df.head(20)
```

```python
# Declare year lists to be converted into columns
y2000 = [2000, 2000, 2000, 2000, 2000, 2000, 2000, 2000, 2000, 2000, 2000, 2000, 2000]
y2004 = [2004, 2004, 2004, 2004, 2004, 2004, 2004, 2004, 2004, 2004, 2004, 2004, 2004]
y2008 = [2008, 2008, 2008, 2008, 2008, 2008, 2008, 2008, 2008, 2008, 2008, 2008, 2008]
y2012 = [2012, 2012, 2012, 2012, 2012, 2012, 2012, 2012, 2012, 2012, 2012, 2012, 2012]
y2016 = [2016, 2016, 2016, 2016, 2016, 2016, 2016, 2016, 2016, 2016, 2016, 2016, 2016]
y2020 = [2020, 2020, 2020, 2020, 2020, 2020, 2020, 2020, 2020, 2020, 2020, 2020, 2020]
```

```python
df_2000 = scaled_df[scaled_df.YEAR == 2000]
df_2004 = scaled_df[scaled_df.YEAR == 2004]
df_2008 = scaled_df[scaled_df.YEAR == 2008]
df_2012 = scaled_df[scaled_df.YEAR == 2012]
df_2016 = scaled_df[scaled_df.YEAR == 2016]
df_2020 = scaled_df[scaled_df.YEAR == 2020]
```

```python
y_2000 = df_2000.OUTCOME
X_2000 = df_2000.drop(columns=['YEAR','STATE','OUTCOME', 'guns_per_household', 'med_house_inc', 'violent_crime_rate', 
                               'population', 'PCT_UNDER_40', 'bachelors_degree_tot'])

y_2004 = df_2004.OUTCOME
X_2004 = df_2004.drop(columns=['YEAR','STATE','OUTCOME', 'guns_per_household', 'med_house_inc', 'violent_crime_rate', 
                               'population', 'PCT_UNDER_40', 'bachelors_degree_tot'])

y_2008 = df_2008.OUTCOME
X_2008 = df_2008.drop(columns=['YEAR','STATE','OUTCOME', 'guns_per_household', 'med_house_inc', 'violent_crime_rate', 
                               'population', 'PCT_UNDER_40', 'bachelors_degree_tot'])

y_2012 = df_2012.OUTCOME
X_2012 = df_2012.drop(columns=['YEAR','STATE','OUTCOME', 'guns_per_household', 'med_house_inc', 'violent_crime_rate', 
                               'population', 'PCT_UNDER_40', 'bachelors_degree_tot'])

y_2016 = df_2016.OUTCOME
X_2016 = df_2016.drop(columns=['YEAR','STATE','OUTCOME', 'guns_per_household', 'med_house_inc', 'violent_crime_rate', 
                               'population', 'PCT_UNDER_40', 'bachelors_degree_tot'])

y_2020 = df_2020.OUTCOME
X_2020 = df_2020.drop(columns=['YEAR','STATE','OUTCOME', 'guns_per_household', 'med_house_inc', 'violent_crime_rate', 
                               'population', 'PCT_UNDER_40', 'bachelors_degree_tot'])
```

```python
# Find 2000 coefficients and store in a DF
logreg = LogisticRegression()
logreg.fit(X_2000, y_2000)
log_odds = logreg.coef_[0]
coef2000_df = pd.DataFrame(log_odds, 
             X_2000.columns, 
             columns=['coef'])\
            .sort_values(by='coef', ascending=False)
coef2000_df
```

```python
# Use 'Year' as the column name and equate it to the corresponding year list
coef2000_df['Year'] = y2000
coef2000_df.reset_index(inplace=True)
coef2000_df
```

```python
# Use 'Year' as the column name and equate it to the corresponding year list
coef2004_df['Year'] = y2004
coef2004_df.reset_index(inplace=True)
coef2004_df
```

```python
# Append all coefficient DFs into 1
coef_long_df = coef2000_df.append(coef2004_df,ignore_index=True)
coef_long_df = coef_long_df.append(coef2008_df,ignore_index=True)
coef_long_df = coef_long_df.append(coef2012_df,ignore_index=True)
coef_long_df = coef_long_df.append(coef2016_df,ignore_index=True)
coef_long_df = coef_long_df.append(coef2020_df,ignore_index=True)
coef_long_df['Party'] = np.where(coef_long_df['coef']>=0, 'Republican', 'Democrat')
coef_long_df = coef_long_df.rename(columns = {'index':'Feature', 'coef':'Coefficient'}, inplace = False)
coef_long_df['Coefficient'] =  coef_long_df['Coefficient'].abs()
coef_long_df
```

```python
# Export the coef_long_df to a csv for visualization use
coef_long_df.to_csv('Resources/coefficients_long_year.csv',index=False)
```

```python
# Transpose the coef_long_df to a short/wide table with all yearly coefficients on one line
coef_short_df = coef_long_df.pivot(index='Year', columns='Feature', values = 'Coefficient')
coef_short_df.reset_index(inplace=True)
coef_short_df
```

```python
# Export the coef_short_df to a csv for visualization use
coef_short_df.to_csv('Resources/coefficients_short_year.csv',index=False)
```
